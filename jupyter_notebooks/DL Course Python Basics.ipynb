# Functions

```python
def print_squares(end_val):
  for i in range(end_val):
    print (i , i**2)
```

```python
print_squares(5)
```

```python
print_squares(3)
```

```python
def sum_of_squares(end_val):
  sum_ = 0
  for i in range(end_val):
    sum_ += i ** 2
  return sum_
```

```python
a = sum_of_squares(5)
print(a)
```

# Class


```python
class StateReview: #provides encyapsulation and code reuse
  """This is a sample class to illustrate how Python classes work"""
  def __init__(self, name = None, is_rainy = None, districts = None):
    self.name = name
    self.is_rainy = is_rainy
    self.districts = districts
    self.rating = -1
    
  def has_rating(self):
    return self.rating > -1
```

```python
new_state = StateReview('Kerala')
```

```python
type(new_state)
```

```python
print(new_state.name, new_state.is_rainy, new_state.districts)
```

```python
new_state.is_rainy = True
new_state.districts = 14
new_state.rating = 4.3
```

```python
new_state.has_rating()
```
